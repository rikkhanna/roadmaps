Scatter Plot


```python
rng = np.random.default_rng(42)
x   = rng.normal(0, 1, 300)
y   = x * 0.6 + rng.normal(0, 0.8, 300)
c   = np.hypot(x, y)      # colour mapped to distance from origin
fig, ax = plt.subplots()
sc = ax.scatter(x, y,
                c=c,              # colour array or single colour
                cmap='plasma',   # see plt.colormaps()
                s=40,             # marker size in points²
                alpha=0.7,
                edgecolors='none')
fig.colorbar(sc, ax=ax, label='Distance')
```

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(10, 6))
gs  = gridspec.GridSpec(2, 3,
                          hspace=0.4,
                          wspace=0.3)

ax_main  = fig.add_subplot(gs[0, :])    # spans all 3 columns in row 0
ax_left  = fig.add_subplot(gs[1, 0:2]) # spans 2 cols
ax_right = fig.add_subplot(gs[1, 2])   # single cell
```

```python
import sys
!{sys.executable} -m pip install seaborn
import seaborn as sns
import pandas as pd

# Correlation matrix heatmap
df = sns.load_dataset('penguins').dropna()
corr = df.select_dtypes('number').corr()

fig, ax = plt.subplots(figsize=(7, 6))
sns.heatmap(corr,
            ax=ax,
            annot=True,           # show values in cells
            fmt='.2f',            # float format string
            cmap='coolwarm',
            vmin=-1, vmax=1,
            center=0,             # 0 = white for diverging maps
            linewidths=0.5,
            square=True,
            cbar_kws={'shrink': 0.8})
```

```python
# Hierarchically clusters rows and columns automatically
g = sns.clustermap(corr,
                    cmap='vlag',
                    standard_scale=1,   # z-score columns
                    figsize=(8, 8),
                    annot=True, fmt='.1f')
g.ax_heatmap.set_title('Clustered Correlation')
```
