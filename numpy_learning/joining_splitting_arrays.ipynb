Joining and Splitting arrays

```python
import numpy as np
```

```python
first_arr = np.arange(1,11)
second_arr = np.arange(11,21)

con_arr = np.concatenate((first_arr, second_arr))
con_arr
```

```python
stack_arr = np.stack((first_arr,second_arr))

stack_arr
```

```python
hstack_Arr = np.hstack((first_arr,second_arr))
hstack_Arr
```

```python
vstack_arr = np.vstack((first_arr, second_arr))
vstack_arr
```

Splitting Arrays

- Splitting breaks one array into multiple subarrays.
- functions -> `split`, `array_split`, `hsplit`, `vsplit`

Major drawback
- Will split only whne number of elements is divisible by number of splits. So the resulting arrays needs to have same shape.

```python
new_arr = np.arange(1,13)
new_arr
```

```python
split_arr = np.split(new_arr, 4)
split_arr
```
