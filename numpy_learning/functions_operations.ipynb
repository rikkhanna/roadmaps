Arithmetic operations and functions

```python
import numpy as np
```

- Numpy has a powerful capacity to apply operations on the color array instead of each element individually. This is called vectorization.

Vectorization operation can be executed in parallel on multiple elements of array.

Benefits
- Higher Performance Code
- Less Verbose Code
- Better maintainability



```python
a=np.arange(1,11)
b=np.arange(21,31)
print("a", a)
print("b", b)
print(a+b)
print(a*b)
b/a
```

```python
np.add(a,b)
```

```python
np.subtract(b,a)
```

```python
np.multiply(a,b)
```

```python
np.divide(b,a)
```

```python
np.mod(b,a)
```

```python
np.power(a,b)
```

```python
np.sqrt(a)
```
