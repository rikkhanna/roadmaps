Intrinsic NumPy array Creation

```python
import numpy as np
```

```python
integer_array = np.arange(10)
integer_array
```

```python
sec_int = np.arange(100,120)
sec_int
```

```python
random_num = np.random.rand(4,4)
random_num
```

```python
random_int = np.random.randint(0,100,4)
random_int
```

Creating Arrays filled with constant values

```python
first_zero_arrays = np.zeros(5)
first_zero_arrays
```

```python
first_ones_array = np.ones(5)
first_ones_array
```

```python
first_ones_multi_dim = np.ones((2,4), dtype=int)
first_ones_multi_dim
```

```python
first_empty_array = np.empty(5)
first_empty_array
```

```python
first_empty_array_type = np.empty(10, dtype=int)
first_empty_array_type
```

```python
first_empty_array_type.fill(12)
first_empty_array_type
```

```python
first_full_array = np.full((2,2), 8, dtype=float)
first_full_array
```
