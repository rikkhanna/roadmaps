Indexing and Slicing


```python
import numpy as np
```

```python
three_dim = np.reshape(np.arange(12), (3,2,2))
three_dim
```

```python
three_dim[0,1,1]
```

```python
three_dim[2,-1,-1]
```

```python
one_dim = np.arange(10)
one_dim
```

```python
one_dim[2:6]
```

```python
one_dim[-1]
```

```python
one_dim[-3:]
```
