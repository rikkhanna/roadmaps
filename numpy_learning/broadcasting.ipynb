Broadcasting
- Sometimes you need to apply arithmetic operations on arrays with different shapes.

- NumPy provides a way to achieve this Broadcasting.
- In cases when arrays have unequal sizes, the arithmentic operations are defined if the small array can be effectively expanded to match the larger array by using simple rules.

Dimensions are compatible
- If their axes are on a one-by-one basis, they have either the same length or a length of one

```python
import numpy as np
```

```python
a=np.arange(1,10).reshape(3,3)
a
```

```python
b=np.arange(1,4)
b
```

```python
a+b
```

```python
c=np.arange(1,3)
c
```

```python
a+c
```

```python
d=np.arange(24).reshape(2,3,4)
d
```

```python
e=np.arange(4)
e
```

```python
d-e
```
