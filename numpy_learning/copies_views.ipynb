Copies and views


```python
import numpy as np
```

Assignment != Copy -> It copies the reference to memory of the numpy object

```python
stud_num = np.array([1,2,3])
stud_num2 = stud_num
print(stud_num2)
stud_num2[1] = 4
print(stud_num2)
print(stud_num)     
```

For copy -> use only copy()

```python
stud_num2 = stud_num.copy()
stud_num2[1] = 20
print(stud_num2)
print(stud_num)
```
