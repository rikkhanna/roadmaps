Adding, removing and sorting elements   

```python
import numpy as np
```

```python
first_arr = np.array([1,2])
new_arr = np.insert(first_arr, 1,4)
new_arr


                    
```

```python
second_arr = np.array([2,4,1])
second_arr = np.append(second_arr, 5)
second_arr
```

```python
second_arr = np.delete(second_arr, 2)
second_arr
```

```python
ints_2dim = np.array([[3,2,5,7], [5,0,8,3]])
ints_2dim
```

```python
print(np.sort(ints_2dim))
```
