Reshaping Arrays


```python
import numpy as np

```

```python
first_arr = np.arange(1,13)
first_arr
```

```python
second_arr = np.reshape(first_arr, (3,4))
second_arr
```

```python
third_arr = np.reshape(first_arr, (2,6))
third_arr
```

```python
fourth_arr = np.reshape(first_arr, (3,2,2))
fourth_arr
```
